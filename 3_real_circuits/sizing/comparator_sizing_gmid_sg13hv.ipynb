# Latched Comparator Sizing using gm/ID (SG13G2 HV Devices)

## Setup & Device Models

```python
import numpy as np
from pygmid import Lookup as lk

nmos = lk('sg13_hv_nmos.mat')
pmos = lk('sg13_hv_pmos.mat')
```

## System Specifications

```python
vdd = 3.0
vcm = vdd / 2           
f_clk = 1e6 #clock frequency

print(f"VDD   = {vdd} V")
print(f"Vcm   = {vcm} V")
print(f"f_clk = {f_clk/1e6:.1f} MHz")
```

## Process Minimum Sizes

```python
# SG13G2 HV process minimum dimensions (µm)
L_min_n = 0.45   
L_min_p = 0.40   
W_min   = 0.30   

print(f"L_min_n = {L_min_n} µm")
print(f"L_min_p = {L_min_p} µm")
print(f"W_min   = {W_min} µm")
```

## gm/ID Design Choices

```python
gmid_in = 10    
```

## Length Choices

```python
L_in      = 1.0       
L_latch_n = L_min_n   
L_latch_p = L_min_p   
L_eval    = L_min_n   
L_reset   = L_min_n   
```

## Operating Point Estimates

```python
vds_in = vdd - vcm       
vsb_in = 0.0 #source = bulk = VDD

print(f"vds_in = {vds_in:.3f} V")
print(f"vsb_in = {vsb_in:.3f} V")
```

## Input Pair Sizing

```python
Id_branch = 3.14e-6 #same as OTA

jd_in = pmos.lookup('ID_W', GM_ID=gmid_in, L=L_in, VDS=vds_in, VSB=vsb_in)
W_in = Id_branch / abs(jd_in)
W_in_round = max(round(W_in * 2) / 2, 0.5) 

gm_in = gmid_in * Id_branch

print(f"JD (PMOS input)  = {abs(jd_in)*1e6:.3f} µA/µm")
print(f"W_in (exact)     = {W_in:.2f} µm")
print(f"W_in (rounded)   = {W_in_round:.1f} µm")
print(f"W·L product      = {W_in_round * L_in:.1f} µm²")
print(f"gm_in            = {gm_in*1e6:.2f} µS")
```

## NMOS Latch & Eval Switch Sizing

```python
ratio = 3.5
W_latch_n = W_in_round / ratio
W_latch_n_round = max(round(W_latch_n * 2) / 2, 0.5)

W_eval = W_latch_n_round #eval switches matched to latch

print(f"W_in / ratio     = {W_in_round} / {ratio} = {W_latch_n:.2f} µm")
print(f"W_latch_n        = {W_latch_n_round:.1f} µm")
print(f"W_eval           = {W_eval:.1f} µm")
print(f"Actual ratio     = {W_in_round / W_latch_n_round:.1f}")
```

## Remaining Devices (Minimum Size)

```python
W_latch_p = W_min   # PMOS cross-coupled latch
W_reset   = W_min   # Reset switches

print(f"W_latch_p (XM3/4)  = {W_latch_p} µm  (minimum)")
print(f"W_reset (XM9/10)   = {W_reset} µm  (minimum)")
```
